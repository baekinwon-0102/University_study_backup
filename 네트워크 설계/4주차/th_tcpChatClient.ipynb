```python
# TCP echo client program

import socket,threading

def recv(client):
    while True:
        rxmsg = client.recv(1024)
        msg = rxmsg.decode('utf-8')
        print(">> Server msg: ",msg)
        if msg == 'exit':
            client.close()
            print(">> Disconnect by Server, Pleas input 'exit'....")
            break

def send(client):
    while True:
        msg = input(">> Client msg: ")
        client.sendall(msg.encode('utf-8'))
        if msg == 'exit':
            print(">> Exit by User")
            break

if __name__ == "__main__":
    IP = 'localhost'
    PORT = 9999
    address = (IP, PORT)
    print(">> Echo client start")
    
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    
    client.connect(address)
    print(">> Connect server")
    
    receiver = threading.Thread(target = recv, args=(client,))
    receiver.start()
    sender = threading.Thread(target = send, args=(client,))
    sender.start()
    while True:
        receiver.join()
        sender.join()
        break
    
    client.close()
    print(">> Client End")
```
