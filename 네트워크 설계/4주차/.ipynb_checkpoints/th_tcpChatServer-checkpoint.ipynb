```python
# TCP Echo server program
import socket, threading

def recv(conn):
    while True:
        
        # recv(받을 메시지크기) 통신소켓에 데이터가 들어올때 까지 기다림
        # 통신소켓에 데이터가 들어오면 메시지 반환
        rxmsg = conn.recv(1024)
        msg = rxmsg.decode('utf-8') # 받을 때 디코딩
        print(">> Client msg:", msg)
        if msg == 'exit':
            # close() 서버 소켓 종료
            print(">> Disconnect by Client, Please input 'exit'....")
            
            break

def send(conn):
    while True:
        msg = input(">>Server msg:")
        conn.sendall(msg.encode('utf-8')) # 보낼때 인코딩
        if msg == 'exit':
            print(">> Exit by User")
            break

if __name__=="__main__":
    IP = 'localhost' # 루프 백 IP
    PORT = 9999
    address = (IP, PORT) # tuple로 주소정보 저장
    print(">> Echo server start")
    
    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    
    server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR,1)
    
    server.bind(address)
    
    server.listen(1) # 클라이언트 하나만 받음
    print(">> Echo server wait..........")
    conn, addr = server.accept()
    print(f">> {addr} is connected")

    # 스레드 객체 생성하고 실행하기
    receiver = threading.Thread(target = recv, args=(conn,))
    receiver.start()
    sender = threading.Thread(target = send, args=(conn,))
    sender.start()
    receiver.join()
    sender.join()

    conn.close()
    server.close()
    print(">> Server End")
```
