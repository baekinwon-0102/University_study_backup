```python
# UDP 소켓을 이용한 파일전송 클라이언트 프로그램
import socket, time
import os, select # 파일을 처리하기 위한 라이브러리
from PIL import Image # 파이썬 이미지 처리 라이브러리

IP = '10.2.40.241'
PORT = 8080
address = (IP,PORT)
timeout = 3 # select 함수에서 사용

print(">> UDP Client Start")
# UDP 소켓 객체 생성
client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# 닉네임 지정
nickname = input("Choose your nickname: ")
# sendto(data,addr) 서버로 메시지(닉네임) 전송
client.sendto(nickname.encode('utf-8'),address)

# recvfrom(): 데이터(파일명) 수신
data, addr = client.recvfrom(1024)
if data:
    file_name = data.decode('utf-8')
    # 디렉토리에 같은 파일명이 있는 경우 파일명 변경
    if os.path.exists(file_name): # 현 디렉토리에 동일 파일명 있다면 동작
        index = file_name.find('.') # 파일명에 있는 온점 찾기
        old_name = file_name[:index] # 온점 기준으로 앞은 이름
        file_extension = file_name[index:] # 온점 기준으로 뒤는 파일 확장자
        file_name = old_name + '_copy' + file_extension # 기존파일명_copy.확장자
    print(">> file name:",file_name)

# 파일 오픈해서 수신된 이미지 저장, 이미지는 바이너리로 저장
f = open(file_name, 'wb')
while True: # 수신루프
    print('.',end ='')
    #소켓에 수신된 파일 있는지 체크하는 함수(멀티플레시 함수)
    ready = select.select([client],[],[],timeout) # 소켓에 데이터가 들어온것이 있는지 
    if ready[0]: # 파일이 있는 경우
        data, addr = client.recvfrom(1024)
        f.write(data)
    else: # 파일이 다 수신된 경우
        print(f"{file_name} Finish!")
        f.close()
        # 수신된 이미지 파일을 화면에 출력
        img = Image.open(file_name)
        print('Image showing ........')
        img.show()
        break
client.close()
print(">> UDP client End")
```
