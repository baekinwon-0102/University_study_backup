```python
# UDP 소켓을 이용한 에코서버 서버 프로그램
import socket

IP = 'localhost'
PORT = 9000
address = (IP,PORT)

print(">> UDP Echo Server start")
# UDP 소켓 객체 생성 (시험 나옴)
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
# bind(): 소켓에 IP 주소와 port 번호 지정
server.bind(address)

while True:
    #recvfrom():클라이언트 메시지 수신
    data, addr = server.recvfrom(1024)
    print(f"{addr}: {data.decode('utf-8')}")

    # sendto(data,addr): 클라이언트가 보낸메시지 그대로 돌려주기
    server.sendto(data,addr)

server.close()
print(">> UDP Echo Server End")
```
