```python
# UDP 소켓을 이용한 에코서버 클라이언트 프로그램
import socket

IP = 'localhost'
PORT = 9000
address = (IP,PORT)

print(">> UDP Echo Server start")
# UDP 소켓 객체 생성
client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

while True:
    # 보낼메시지 입력
    sendmsg = input("Send: ")
    # sendto() 서버로 메시지 보내기
    if sendmsg == 'exit':
        print("Disconnect by User")
        break
    client.sendto(sendmsg.encode('utf-8'),address)
    # recvfrom() 서버로부터 되돌아 온 메시지 수신
    data, addr = client.recvfrom(1024)
    print("Recv:",data.decode('utf-8'))

client.close()
print(">> UDP Echo client End")
```
