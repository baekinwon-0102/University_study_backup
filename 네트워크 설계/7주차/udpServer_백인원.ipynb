```python
# 파일 입출력하기
# 파일 객체 만들고 쓰기(쓰기모드로 파일 열어서 데이터 저장)
# open(): 파일 열고 파일 객체 반환(read/write)
f = open('text.txt','w',encoding='utf-8')
f.write("백인원님\n") # 파일에 값 입력
f.write("안녕하세요\n")
f.write("반갑습니다\n")
f.close() # 파일 닫기
```

```python
# 파일 객체 만들고 읽기(읽기모드로 파일 열어서 데이터 읽기)
# read(): 파일 전체 내용 한번에 읽어서 저장
# readline(): 파일 내용 한 줄씩 읽어서 저장
# readlines(): 파일 내용 모두 읽어 리스트에 저장
f = open('text.txt','r',encoding='utf-8')
data = f.read() # 파일전체를 문자열로 리턴
print(data)
f.close()
```

```python
# UDP 소켓 이용한 파일전송 서버 프로그램
import socket, time

IMAGE_FILE = 'intro.png'

IP = 'localhost'
PORT = 8888
address = (IP,PORT)

file_name = IMAGE_FILE
print("file:",file_name)

print(">> UDP Server start")
# UDP 소켓 객체 생성
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
# bind(): 소켓에 IP주소와 port번호 지정
server.bind(address)

# recvfrom(): 클라이언트 메시지(닉네임) 수신
data, addr = server.recvfrom(1024)
print(f"{data.decode('utf-8')}, {addr}님이 접속하였습니다")

# sendto(data,addr): 클라이언트 파일이름 전송
server.sendto(file_name.encode('utf-8'),addr)
print(f"Sending {file_name} .....") # file_name 전송

# 파일 열고 이미지 데이터 전송, 이미지는 바이너리로 지정
f = open(file_name, 'rb')
data = f.read(1024) # 1KB씩 읽어서 전송

while data:
    # 파일 전송, 바이너리인경우 인코딩 필요 없음
    if server.sendto(data, addr): # 클라이언트로 메시지를 성공적으로 보냈다면 동작
        data = f.read(1024) # 1KB씩 읽어서 전송
        time.sleep(0.02) # 클라이언트 데이터 저장시간 보장

server.close() # 서버 소켓 닫기
f.close() # 파일 닫
print(">> UDP Server End")
```
