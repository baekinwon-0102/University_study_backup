```python
# 스레드간 통신하기

import threading, time

def total_sum(start,end,sumList):
    result = 0
    for i in range(start, end):
        result+=i
    sumList.append(result)

if __name__ == "__main__":
    start_time = time.time() # 시작시간
    numbers = list()
    start = 0
    end = 100000000

    t = threading.Thread(target=total_sum, args=(start,end+1,numbers))
    t.start()
    t.join()
    print(f"결과:{sum(numbers)}, 소요시간: {time.time()-start_time}")
```

```python
# 스레드간 통신하기
# 방법1: list 사용하기

import threading, time

def total_sum(start,end, sum_list):
    result = 0
    for i in range(start, end):
        result+=i
    sum_list.append(result)

if __name__ == "__main__":
    start_time = time.time() # 시작시간
    numbers = list()
    start = 0
    end = 100000000

    t1 = threading.Thread(target=total_sum, args=(start,end//2,numbers))
    t2 = threading.Thread(target=total_sum, args=(end//2,end+1,numbers))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    
    print("total sum:",sum(numbers), "소요시간:", time.time()-start_time)
```
