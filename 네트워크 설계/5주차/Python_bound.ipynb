```python
# 파이썬 멀티스레드의 성능과 GIL

import threading, time

# CPU_bound job: CPU연산을 많이 사용하는 작업 (CPU 활용률이 많음)
def cpu_bound(n):
    current_thread = threading.current_thread().name # 현재 돌고있는 스레드 이름 알려줌
    print(f"{current_thread} job start")
    while n:
        n -=1
    print(f"{current_thread} job end")
if __name__ == "__main__":
    start = time.time() # 시작 시간
    n = 100000000
    cpu_bound(n)
    cpu_bound(n)
    print("cpu_bound 소요시간:",time.time()-start)
```

```python
# 파이썬 멀티스레드의 성능과 GIL

import threading, time

# CPU_bound job: CPU연산을 많이 사용하는 작업 (CPU 활용률이 많음)
def cpu_bound(n):
    current_thread = threading.current_thread().name # 현재 돌고있는 스레드 이름 알려줌
    print(f"{current_thread} job start")
    while n:
        n -=1
    print(f"{current_thread} job end")
if __name__ == "__main__":
    start = time.time() # 시작 시간
    n = 100000000
    t1 = threading.Thread(target=cpu_bound, args=(n,))
    t2 = threading.Thread(target=cpu_bound, args=(n,))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    
    print("cpu_bound 소요시간:",time.time()-start)
```

```python
# 파이썬 멀티스레드의 성능과 GIL

import threading, time

# IO_bound job: IO 연산을 많이 사용하는 작업 (CPU 활용률이 적음)
def io_bound(n):
    current_thread = threading.current_thread().name # 현재 돌고있는 스레드 이름 알려줌
    print(f"{current_thread} io_bound job start")
    time.sleep(n)
    print(f"{current_thread} io_bound job end")
if __name__ == "__main__":
    start = time.time() # 시작 시간
    n = 5
    io_bound(n)
    io_bound(n)
    print("io_bound 소요시간:",time.time()-start)
```
