```python
# TCP echo client program

import socket

IP = ''
PORT = 1234
address = (IP, PORT)
print(">> Echo client start")

client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

client.connect(address)
print(">> Connect Game Server")
while True:
    msg = input(">> 1-9사이의 숫자를 입력하세요(0은 게임포기): ")
    client.sendall(msg.encode('utf-8'))

    rxmsg = client.recv(1024)
    msg2 = rxmsg.decode('utf-8')
    print(msg2)

    # if msg == '0':
    #     print(">> Client End")
    #     break
        
    if msg2 == '정답':
        print(">> Client End")
        break
    
client.close()
```
