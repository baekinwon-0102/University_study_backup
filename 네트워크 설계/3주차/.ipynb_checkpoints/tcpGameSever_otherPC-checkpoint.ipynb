```python
# TCP Echo server program
import socket
import random

IP = '' # 루프 백 IP
PORT = 8888
address = (IP, PORT) # tuple로 주소정보 저장

print(">> Game server start")

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# (optional code) winError 10048 에러 해결을 위한 추가 코드
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR,1)

# bind(주소정보) 소켓에 IP주소와 포트번호 지정
server.bind(address)

# listen() 클라이언트 접속요청 대기 상태
server.listen(1) # 클라이언트 하나만 받음
print(">> Game server wait..........")
conn, addr = server.accept()
print(f">> {addr} is connected")
# accpet() 클라이언트 접속 요청이 있을 시 수락
# 통신소켓과 클라이언트 주소정보를 반환
random_key = random.randrange(1,1001)
print('Server Key value is ',random_key)
while True:
    rxmsg = conn.recv(1024)
    msg = rxmsg.decode('utf-8')
    print(">> recv number: ", msg)
    # if msg.isdigit()==False:
    #     ans = '입력값이 잘못되었습니다.'
    #     conn.sendall(ans.encode('utf-8'))
    #     continue
    
    # if int(msg) == 0:
    #     ans = '>> Answer:'
    #     conn.sendall(ans.encode('utf-8'))
    #     print(">> Disconnect by Client")
    #     print(">> Game Server End")
    #     break
    if int(msg) > random_key:
        ans = '낮음'
        conn.sendall(ans.encode('utf-8'))
    elif int(msg) < random_key:
        ans = '높음'
        conn.sendall(ans.encode('utf-8'))
    elif int(msg) == random_key:
        ans = '정답'
        conn.sendall(ans.encode('utf-8'))
        print(">> Client has been correct Answer")
        print(">> Game Server End")
        break
    else:
        continue
    
# close() 서버 소켓 종료
conn.close() # 통신 소켓 종료
server.close()  # 리스닝 소켓 종료
```
