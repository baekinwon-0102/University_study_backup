```python
import socket
# socket(): 소켓 생성(리스닝 소켓), bind(): 소켓에 IP주소와 포트번호 할당, listen(): 클라이언트와 연결 대기, accept(): 연결수락 및 새로운 통신소켓 생성
# sendall()/recv(): 메시지 송수신(통신소켓)
# close(): 소켓 닫기
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # AF_INET = IP주소 형식(IPv4), SOCK_STREAM = 통신방식(TCP)
                                                           # AF_INET6 = IPv6, SOCK_DGRAM = UDP, SOCK_RAW= IP프로토콜 직접 사용해 통신할 때 사용
print(server) #family = IP버전, type = TCP를 쓸지 UDP를 쓸지 정함
```

```python
# TCP Echo server program
import socket

IP = 'localhost' # 루프 백 IP
PORT = 9999
address = (IP, PORT) # tuple로 주소정보 저장

print(">> Echo server start")

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# (optional code) winError 10048 에러 해결을 위한 추가 코드
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR,1)

# bind(주소정보) 소켓에 IP주소와 포트번호 지정
server.bind(address)

# listen() 클라이언트 접속요청 대기 상태
server.listen(1) # 클라이언트 하나만 받음
print(">> Echo server wait..........")
conn, addr = server.accept()
print(f">> {addr} is connected")
# accpet() 클라이언트 접속 요청이 있을 시 수락
# 통신소켓과 클라이언트 주소정보를 반환
while True:
    
    # recv(받을 메시지크기) 통신소켓에 데이터가 들어올때 까지 기다림
    # 통신소켓에 데이터가 들어오면 메시지 반환
    rxmsg = conn.recv(1024)
    msg = rxmsg.decode('utf-8') # 받을 때 디코딩
    print(">> Server msg:", msg)
    if msg == 'exit':
        # close() 서버 소켓 종료
        conn.close() # 통신 소켓 종료
        server.close()  # 리스닝 소켓 종료
        print(">> Disconnect by Client")
        
        break
    # sendall(보낼데이터) 통신소켓에 메시지 보내기
    # 에코서버는 받은 데이터 그대로 돌려줌
    msg = input(">>Client msg:")
    conn.sendall(msg.encode('utf-8')) # 보낼때 인코딩
    if msg == 'exit':
        # close() 서버 소켓 종료
        conn.close() # 통신 소켓 종료
        server.close()  # 리스닝 소켓 종료

        print(">> Exit by User")
        break
    

```
