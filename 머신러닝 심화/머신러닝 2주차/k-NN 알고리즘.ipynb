```python
from sklearn import datasets

iris = datasets.load_iris()

x = iris.data[:,:]# 학습 데이터 [:,:]는 2개의 차원의 모든 값을 가져옴 
y = iris.target # 타겟 데이터
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 30)

clf = KNeighborsClassifier(n_neighbors=11)
clf.fit(x_train, y_train)
#prediction = clf.predict(x_test)
Test_x = [[5.7,2.8,3.7,1.4]]
#Test_x = [[5.7,2.8]]
prediction = clf.predict(Test_x)

print(Test_x, iris['target_names'][prediction])

#print("clf. score            : {0:.3f}".format(clf.score(x_train,y_train)))
#print("(pred == y_test) score: {0:.3f}".format((prediction==y_test).mean()))
```

```python
import pandas as pd
import mglearn
iris_df = pd.DataFrame(x_train, columns = iris.feature_names)

pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(12,12),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(x_train, y_train)
k3 = [[[4.5,4.5,1.1,0.4]],[[5.3,3.0,4.1,1.2]],[[7.5,3.6,6.2,2.2]]]

for i in range(3):
    prediction = clf.predict(k3[i])
    print(k3[i], iris['target_names'][prediction])
```
